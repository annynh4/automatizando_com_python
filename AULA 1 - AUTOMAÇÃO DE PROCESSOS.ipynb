# Desafio

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadoria da empresa. O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos comprados e o preço médio dos produtos.

E-mail do seu chefe: para o exercício (hashtagtreinamentos@gmail.com)

Link do sistema da empresa: (https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema)

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado.


```python
import pyautogui # Bibilotea de automação
import time 

        # pyautogui.click (para clicar)
        # pyautogui.write(para escrever)
        # pyautogui.press(para apertar/presciona uma única tecla)
        # pyautogui.hotkey(atalho, ex: ctrl + t.)
        # .PAUSE Significa que em todos os comando do pyautogui ele vai esperar (x) seg


pyautogui.PAUSE = 1 

# Passo 1: Acessa o sistema da empresa

pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

        # time.sleep(x) Pra que depois do ENTER ele espere (x) seg
        # a posição do mouse (x=, y=)
        
# Passo 2: Fazer login no sistema

pyautogui.click(x=929, y=480) 
pyautogui.write("anni_nh4")

time.sleep(1)

pyautogui.click(x=948, y=569)
pyautogui.write("05052004")

time.sleep(2)

pyautogui.click(x=922, y=646)

# Passo 3: Baixar a base de dados

        # pyautogui.hotkey("ctrl", "Tab")
        # button=right é pra clicar no lado direito do mouse

time.sleep(4)

pyautogui.click(x=1667, y=458)
```

```python
# Passo 4: Calcular os indicadores

    # pandas  abriviado para pd, é um pacote de código que trabalha enficientemente com base de dados
        
import pandas as pd

            # importar a base de dados
            # antes no link é bom colocar um (r) pra que ele entenda que é um formato RAW = CRU, formato cru para que não venha ocorrer erros.
            # (sep=";") diz pro pd quie o separador dele é o ; e ai vai pegar essa coluna evai separar corretamente.

tabela = pd.read_csv(r"C:\Users\anara\Downloads\Compras.csv", sep=";")
display(tabela)


        # calculo dos indicadores
        # .sum para somar
        # parta selecionar uma coluna usamos colchetes []

    # total gasto -> soma a coluna valor final
    
total_gasto = tabela["ValorFinal"].sum()

    # quantidade - > somar a coluna quantidade

quantidade = tabela["Quantidade"].sum()

    # preço médio - > total gasto / quantidade

preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)

```

```python
import pyperclip

# Passo 5: Enviar o email para a diretoria/para o chefe

pyautogui.PAUSE = 1

        # entrar no email
    
pyautogui.hotkey("ctrl", "t")
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")

time.sleep(5)

pyautogui.click(x=101, y=254) # clilcar em ESCREVER

time.sleep(3)

pyautogui.write("anabraidz@gmail.com")
pyautogui.press("tab") #SELECIONA O EMAIL

pyautogui.press("tab") # PASSA PRO CAMPO DO ASSUNTO
pyperclip.copy("Relatório de Compras") # Título do emaik
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") #PASSAR PRO CAMPO DO CORPO DO EMAIL
# o python entender que ao por 3 aspas se torna um texto

# para caracteres especiais usamos o pyperclip que permite copiar um texto que tá no código e colar ele com pyautogui.

# um f antes do texto ele formata.

# código de formatação:

# dois pontos (:) significa que eu quero formatar;
# vírgula (,) é um separador de milhar;
# (.2f) significa que são duas casas decimais.
# ponto final (.) significa que tem casa decimal;
# Número dois (2) quantidade de casa decimal;
# letra (f) pra dizer que é um (float) que é um número que tem casa decimal.

texto = f"""
Prezados,

Segue o relatório de compras,

Total gasto: R${total_gasto:,.2f}

Quantidade: R${quantidade:,.2f}

Preço Médio: R${preco_medio:,.2f}

Qualquer dúvida é só me retornar.

Att,
Ana Carolina Ramos Azevedo

Santander Financiamentos

OBS: TESTE DE AUTOMAÇÃO

HASHTAG TREINAMENTOS

"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar

pyautogui.hotkey("ctrl", "enter")

time.sleep(2)

pyautogui.click (x=507, y=879) # para abrir o email enviado
```
